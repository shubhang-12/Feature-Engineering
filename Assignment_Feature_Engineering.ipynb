1. What is a parameter?
 - In machine learning, a parameter is an internal variable within a model that the algorithm learns from the training data. These values define how the model transforms input data into an output and are adjusted automatically during the training process. They are integral to the model's structure and are optimized during training, unlike hyperparameters which are set beforehand.

2. What is correlation? What does negative correlation mean?
 - Correlation is a statistical measure that expresses the extent to which two variables are related. It doesn't necessarily imply causation, but it indicates the strength and direction of their association. For example, you might find a correlation between hours of study and exam scores, suggesting that as study time increases, exam scores tend to increase as well.

 A negative correlation, also known as an inverse correlation, indicates that two variables tend to move in opposite directions. When one variable increases, the other tends to decrease, and vice versa. For instance, the amount of sleep someone gets and their overall stress levels might be negatively correlated, with more sleep potentially leading to less stress.

16. What is an Optimizer? What are different types of optimizers? Explain each with an example.
 - An optimizer is an algorithm used to adjust the parameters (like weights and biases) of a model during training to minimize a loss function. Different optimizers use various strategies to find the optimal parameter values, which helps the model learn from data and make accurate predictions.
 Types of Optimizers and Examples:
 1. **Gradient Descent**:   
This is a fundamental optimization algorithm that iteratively adjusts parameters in the direction of the negative gradient of the loss function.    
Example: Imagine you're on a mountain and want to find the lowest point (minimum of the loss function). Gradient descent would take you downhill in the direction of the steepest slope, iteratively adjusting your position until you reach the lowest point.    
 2. **Stochastic Gradient Descent (SGD)**:   
SGD is a variant of gradient descent that uses a random subset of the training data to calculate the gradient at each iteration.     
Example: Instead of using the entire dataset at once to calculate the gradient, SGD uses a smaller batch of data, making it computationally more efficient and faster to converge.     
 3. **Adam (Adaptive Moment Estimation)**:   
Adam is a popular optimizer that combines the advantages of both SGD with Momentum and RMSProp. It uses adaptive learning rates for each parameter, adjusting them based on the first and second moments of the past gradients.     
Example: Adam can adjust the step size dynamically, taking larger steps when it's far from the optimal point and smaller steps as it gets closer, preventing overshooting.     
 4. **RMSProp (Root Mean Square Propagation)**:    
RMSProp is another adaptive learning rate optimizer that adapts the learning rate based on the magnitude of the gradients.     
Example: RMSProp keeps track of a moving average of the squared gradients, allowing it to adjust the learning rate for each parameter independently.
 5. **Adagrad (Adaptive Gradient)**:   
Adagrad is an adaptive learning rate optimizer that adapts the learning rate based on the historical gradients.     
Example: Adagrad makes the learning rate larger for parameters that are updated infrequently and smaller for parameters that are updated frequently, which can be useful for sparse data.

18. What does model.fit() do? What arguments must be given?
 - The model.fit() function is the core of training a machine learning model in Keras. It's where you feed your training data to the model and initiate the learning process. This function adjusts the model's weights and biases iteratively to minimize the difference between its predictions and the actual target values.    
 Essential Arguments for model.fit():   
 **x**: This is your training data, the input features that the model will learn from. It can be a NumPy array or a TensorFlow Dataset.

 **y**: This represents the target values or labels corresponding to your training data. It should have the same number of samples as x.

 **batch_size**: This determines how many samples are processed at once during training. A smaller batch size might lead to slower training but can improve generalization.

 **epochs**: This specifies the number of times the entire training dataset is passed through the model during training. More epochs can lead to better learning but also increase the risk of overfitting.

 **validation_data**: This is an optional argument where you provide a separate dataset to evaluate the model's performance during training. It helps you monitor overfitting and assess generalization. You can give this argument as a tuple (x_val, y_val) or a TensorFlow Dataset.

 For Example-
 from sklearn.linear_model import LinearRegression
 # Create a model
 model = LinearRegression()
 # Train the model
 model.fit(X_train, y_train)     
 In this example, X_train would be your training data features and y_train would be the corresponding target values.

21. What is feature scaling? How does it help in Machine Learning?
 - Feature scaling is a data preprocessing technique that transforms the values of numerical features in a dataset to a common scale, like 0 to 1 or -1 to 1. It's crucial in machine learning because it prevents features with larger ranges from dominating those with smaller ranges, ensuring all features contribute equally to the model. This leads to improved model performance, faster convergence, and more accurate predictions, especially for algorithms that rely on distance calculations or gradients.
